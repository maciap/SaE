```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time 
import sys
import os
sys.path.append(os.path.abspath(".."))
from discover_near_rank_one_submatrix import SamplingAlgorithm

# Generate points along a line through the origin
n_aligned = 50
t = np.linspace(-10, 10, n_aligned)
line_direction = np.array([1, 2, 3])
aligned_points = np.outer(t, line_direction) + np.random.normal(scale=0.001, size=(n_aligned, 3))

# Generate random points not aligned with the line
n_random = 150
random_points = np.random.uniform(-15, 15, size=(n_random, 3))

# Plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Aligned points
ax.scatter(aligned_points[:, 0], aligned_points[:, 1], aligned_points[:, 2], color='red', label='Aligned Points', s=40)

# Random points
ax.scatter(random_points[:, 0], random_points[:, 1], random_points[:, 2], color='blue', label='Random Points', alpha=0.6)

ax.set_title("3D Data with Subset Aligned to a Line through the Origin", fontsize=14)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()
ax.view_init(elev=20, azim=30)
plt.tight_layout()
plt.show()

# Finally stack random points and line points 
D = np.vstack((aligned_points, random_points)) 
```

```python
u,s,v = np.linalg.svd( aligned_points )
u_random,s_random,v_random = np.linalg.svd( random_points )


# Create figure and two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5), dpi=100)

# First plot
ax1.plot(s, color='tab:blue', linewidth=4)
ax1.set_title("Near-Rank-1 Submatrix", fontsize=14)
ax1.set_xlabel("Index i", fontsize=12)
ax1.set_ylabel(r"Singular value $\sigma_i$", fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.6)

# Second plot
ax2.plot(s_random, color='tab:green', linewidth=4)
ax2.set_title("Background", fontsize=14)
ax2.set_xlabel("Index i", fontsize=12)
ax2.set_ylabel(r"Singular value $\sigma_i$", fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.6)

# Adjust layout
plt.tight_layout()
plt.show()

```

```python
# Run SampleAndExpand with default parameters 
delta = 0.05 # tolerance 
delta_rectangle = 1e-10
SaE_rankone = SamplingAlgorithm(D, delta=delta, delta_rectangle=delta_rectangle, 
approximate_biclique = False, sparsity_constraint = False)
start_time = time.time() 
output =  SaE_rankone.run_multiple(N_rep = 3)
output_submatrix_approximation = output[0]
end_time = time.time() - start_time 
print(f"The output submatrix has dimensions {output[0].shape}")
print(f"Time elapsed {end_time}")
```

```python
best_indices = output[2] # Discovered Submatrix Rank-1 Approximation 
best_D_sub = D[np.ix_(best_indices[0], best_indices[1])] # Discovered Submatrix Original Values
best_error = output[1]
SaE_rankone.plot_approximation(output_submatrix_approximation, best_D_sub, best_error) 
```
